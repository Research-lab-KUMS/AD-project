```python
def truncate(input):
    sr=10000
    sh_result = []
    for i in range(len(input)):
        idx=np.where((input[i] > 60 * sr) & (input[i] < 16* 60 * sr) )[0]
        sh_result.append(input[i][idx])
    return sh_result
```
